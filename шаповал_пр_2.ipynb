```python
# завдання 2
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96#:~:text=%D0%97%D0%B0%20%D0%BF%D1%96%D0%B2%D1%80%D0%BE%D0%BA%D1%83%202023%20%D0%B2%20%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96,%D1%81%D0%BF%D0%B0%D0%B4%20%D0%B7%D0%B0%20%D0%B2%D0%B5%D1%81%D1%8C%20%D1%87%D0%B0%D1%81%20%D0%BD%D0%B5%D0%B7%D0%B0%D0%BB%D0%B5%D0%B6%D0%BD%D0%BE%D1%81%D1%82%D1%96.'
df = pd.read_html(url, thousands=".", decimal=",")[17]
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
df
```

```python
df = df.iloc[:-1]
df.tail(5)
```

```python
avg = df['2008'].mean()
df.loc[df['2008'] > avg]
```

```python
max_value_region = df[df['1990'] == df['1990'].max()]['регіон']
max_value_region
```

```python
max_value_region_1 = df[df['2012'] == df['2012'].max()]['регіон']
max_value_region_1
```

```python
sns.set_style('darkgrid')

plt.bar(df['регіон'], df['2014'])
plt.xlabel("Регіон")
plt.ylabel("Значення")
plt.title("Графік народжуваності в Україні у 2014 році", fontsize=15)

plt.tick_params(labelrotation = 90)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df.drop(['регіон'], axis=1))
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.title('Розподіл коефіцієнта народжуваності в різних регіонах')
plt.show()
```
